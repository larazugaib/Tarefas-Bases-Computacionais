Faça um gráfico da taxa de analfabetismo em São Paulo em função do tempo (anos).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
analfabetismo = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/analfabetismo.csv')
analfabetismo
```

```python
plt.figure()
plt.plot(analfabetismo['Ano'], #abscissa
         analfabetismo['São Paulo']) #ordenada
plt.grid()
plt.title('Taxa De Analfabetismo em São Paulo pelo tempo')
plt.show()
```
